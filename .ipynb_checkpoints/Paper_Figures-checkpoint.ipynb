```python
def fig_label_creator(axs, my_char=ord('a'), c_x = 0, c_y = 0, x_adj = 0, y_adj = 0, fontsize=12, box_color = 'w'):
    j=0
    for ax in axs:
        t = ax.text(c_x,c_y,f'({chr(my_char+j)})', ha='center', va='center', fontsize=fontsize, alpha=0)
        bb = t.get_window_extent().transformed(ax.transData.inverted())
        # rect = plt.Rectangle((bb.x0-0.3, bb.y0+0.3), bb.width, bb.height-1.4, fc='w', alpha=1, zorder=2)
        # ax.add_patch(rect)
        rect = plt.Rectangle((bb.x0-x_adj, bb.y1-y_adj), bb.width+x_adj*2, -bb.height+y_adj*2.2, fc=box_color, zorder=100)#, alpha=1, zorder=2)
        ax.add_patch(rect)
        ax.text(
            rect.xy[0] + rect.get_width() / 2,  # X-center
            rect.xy[1] + rect.get_height() / 2,  # Y-center
            f'({chr(my_char+j)})',  # Text to display
            ha='center', va='center',  # Align text
            fontsize=fontsize, color='black',
            bbox=dict(facecolor='none', edgecolor='none', pad=0),
            zorder=101
        )
        j+=1
        
```

```python
import matplotlib.pyplot as plt

# Create subplots with shared axes
fig, ax = plt.subplots(1, 3, figsize=(10, 10/3), sharex=True, sharey=True)
rnd_no = np.random.default_rng(seed=5)

boundary = boundary_defn.BoundaryCreator(span_x, span_z, del_x_boundary, del_z_spatial, n_layers)

# Plot (a) Uniform
random_generator_settings_dict = {'generator_option': 'uniform', 'max_del_z_per_m': 2}
boundary.generating_boundary(random_generator_settings_dict, rnd_no=rnd_no)
boundary.boundary_init_points('equidistant')
boundary.plot_boundary(ax=ax[0])
ax[0].set(xlabel=None, ylabel=None)
# Plot (b) Normal
random_generator_settings_dict = {'generator_option': 'normal', 'std': 0.2}
boundary.generating_boundary(random_generator_settings_dict, rnd_no=rnd_no)
boundary.boundary_init_points('equidistant')
boundary.plot_boundary(ax=ax[1])
ax[1].set(ylabel=None)
ax[1].set(xlabel=None, ylabel=None)

# Plot (c) FBM
random_generator_settings_dict = {'generator_option': 'fbm', 'H': 0.75, 'length': 1, 'method': 'daviesharte'}
boundary.generating_boundary(random_generator_settings_dict, rnd_no=rnd_no)
boundary.boundary_init_points('equidistant')
boundary.plot_boundary(ax=ax[2])
ax[2].set(xlabel=None, ylabel=None)

# Set shared labels
fig.text(0.5, 0.04, 'Horizontal Distance (m)', ha='center', fontsize=12)
fig.text(0.04, 0.5, 'Depth (m)', va='center', rotation='vertical', fontsize=12)
fig_label_creator(axs=ax, my_char=ord('a'), c_x = 0.2, c_y = 0.07, x_adj = 0, y_adj = 0.04, fontsize = 12, box_color='w')

plt.tight_layout(rect=[0.05, 0.05, 1, 1])  # Leave space for shared labels
plt.show()

```
